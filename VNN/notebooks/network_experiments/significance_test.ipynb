```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy as sp
from scipy.stats import norm

%matplotlib inline
plt.rcParams['figure.figsize'] = [17, 8]
```

```python
pvalues = []
increases = []

def test(standard, comparables):
    standard = np.array(standard).reshape((6,-1)).sum(-1)
    for name in comparables:
        comparable = comparables[name]
        comparable = np.array(comparable).reshape((6,-1)).sum(-1)
        print(name)
        
        _, pvalue = sp.stats.ttest_ind(standard, comparable, equal_var = True)
        pvalues.append(pvalue)
        print(f'\tp = {pvalue:.6f}'.replace('.', ','))

        mean_s = np.mean(standard)
        mean_c = np.mean(comparable)
        increase = (mean_c - mean_s) / mean_s * 100
        increases.append(increase)
        print(f'\tmean: {increase:+.2f}%'.replace('.', ','))
```
